```python
# Import Libraries

import pandas as pd
import numpy as np
import pickle
from TweetProcessor import tweet_processor
from pysentimiento import SentimentAnalyzer
```

```python
# Fetching Data Set

opp = pd.read_csv('/content/drive/MyDrive/Dissertation/Data_Samples/Opposition.csv', index_col=0)
opp.head()
```

```python
# Creating Corpus Object

corpus = list(opp['tweet'])
corpus_cleaned = tweet_processor(corpus)
```

```python
# Creating Sentiment Analyser object

analyzer = SentimentAnalyzer(lang="en")
```

```python
# Fetching probabilities

prob_pos = []
prob_neg = []
prob_nue = []
max_prob = []
loading = 0

for text in corpus_cleaned:
  output = analyzer.predict(text).probas
  prob_pos.append(output['POS'])
  prob_neg.append(output['NEG'])
  prob_nue.append(output['NEU'])
  max_prob.append(max(output,key=output.get))
  loading += 1
  if loading % 1000 == 0:
    print(f"{loading}/{len(corpus_cleaned)}")
```

```python
# DataFrame for Sentiment Analysis Results

opp_sa = pd.DataFrame({'con_id':opp['conversation_id'], 'tweet_cleaned':corpus_cleaned, 'sentiment':max_prob,
                       "POS_Prob":prob_pos, "NEG_Prob":prob_neg,"NEU_Prob":prob_nue})

opp_sa.to_csv("/content/drive/MyDrive/Dissertation/Sentiment Analysis Files/OppSentimentAnalysis.csv")
```
